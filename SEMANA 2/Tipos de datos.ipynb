# Tipos de datos: 

**int**: representa numeros enteros. Ejm: 2,33,0,42,-139,-89

```
type(42)
```

**float**: representa numeros con punto decimal. Ejm: 0.5, 5.0 ,1.25 ,12.0

```
type(0.5)
```

**str**: representa secuncias o cadenas (string) de caracteres. Ejm: "J", "4", "Tengo 10 amigos".

```
type("Aprendo a programar")
```

```
type("4")
```

**bool**: representa valores booleanos o logica binaria: VERDADERO o FALSO. Ejm: True, False.

```
type(False)
```
